```python
'''

There are a few things you can do to increase the comparison between different figures on the same graph:

Use the same axes: This means using the same scale for the x-axis and the y-axis for all of the figures.
                   This will make it easier to see how the different figures compare to each other.
Use the same colors: This will help to visually connect the different figures and make it easier to see how they relate to each other.
Use the same line styles: This is similar to using the same colors, but it can be helpful to use different line styles for different types of data.
                          For example, you might use solid lines for continuous data and dashed lines for discrete data.
Use legends: This is a good way to identify the different figures and to make it clear what each figure represents.
Use annotations: This is a way to add text to the graph to provide additional information about the data.
                 For example, you might use annotations to identify specific points on the graph or to provide explanations of the data
'''
# for example
import matplotlib.pyplot as plt

# Create the first figure
fig1, ax1 = plt.subplots()
ax1.plot([1, 2, 3], [4, 5, 6], color='blue', linestyle='solid')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

# Create the second figure
fig2, ax2 = plt.subplots()
ax2.plot([1, 2, 3], [7, 8, 9], color='red', linestyle='dashed')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

# Add a legend
plt.legend(['Figure 1', 'Figure 2'], loc='upper right')

# Add annotations
ax1.annotate('This is point A', (1, 4), color='black', size='10')
ax2.annotate('This is point B', (2, 8), color='black', size='10')

# Show the graphs
plt.show()
```

```python
# Q6. Before you buy a home, which numpy function will you use to measure your monthly mortgage payment?
```

```python
'''
To calculate the monthly mortgage payment, you typically need to use a formula that takes into account the loan amount,
interest rate, loan term, and other factors. Here's an example of how you can calculate the monthly mortgage payment
using the formula for a fixed-rate mortgage:'''
import numpy as np

def calculate_monthly_payment(loan_amount, interest_rate, loan_term):
    """
    Calculates the monthly mortgage payment for a fixed-rate mortgage.

    Args:
        loan_amount (float): The total loan amount.
        interest_rate (float): The annual interest rate (in decimal form).
        loan_term (int): The loan term in months.

    Returns:
        float: The monthly mortgage payment.
    """
    monthly_interest_rate = interest_rate / 12
    num_payments = loan_term

    # Calculate the monthly payment using the formula
    monthly_payment = (loan_amount * monthly_interest_rate) / (1 - (1 + monthly_interest_rate) ** -num_payments)

    return monthly_payment

# Example usage
loan_amount = 200000  # $200,000
interest_rate = 0.05  # 5% annual interest rate
loan_term = 360       # 30-year loan term

monthly_payment = calculate_monthly_payment(loan_amount, interest_rate, loan_term)
print(f"The monthly mortgage payment is: ${monthly_payment:.2f}")
```
